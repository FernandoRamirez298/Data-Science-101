```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as Rectangle
import datetime as dt
import seaborn as sns
```

```python
df_D= pd.read_csv('../../C05/Data/D.csv')
df_DUK= pd.read_csv('../../C05/Data/DUK.csv')
df_EXC= pd.read_csv('../../C05/Data/EXC.csv')
df_NEE= pd.read_csv('../../C05/Data/NEE.csv')
df_SO= pd.read_csv('../../C05/Data/SO.csv')


df_D["Datafreme"] = "D"
df_DUK["Datafreme"] = "DUK"
df_EXC["Datafreme"] = "EXC"
df_NEE["Datafreme"] = "NEE"
df_SO["Datafreme"] = "SO"

df = [df_D,df_DUK,df_EXC,df_NEE,df_SO]



```

```python
df_final = pd.concat(df)
df_final
```

```python
df_final["Volatility"] = (df_final["High"] - df_final["Low"])/ df_final["Open"]

df_final["Return index"] = (df_final["Close"] / df_final["Open"])-1 


df_final
```

```python
sns.histplot(y=df_final["Volatility"], x= df_final["Date"], hue= df_final["Datafreme"], multiple= "stack")
```
