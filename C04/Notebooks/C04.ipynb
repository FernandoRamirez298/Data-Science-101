```python
df_lluvias= pd.read_csv('../../C04/Data/pune_1965_to_2002.csv')
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop('Year',axis='columns')
df_lluvias
```

```python
x= df_lluvias.index
y= df_lluvias.sum(axis='columns')
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='Precipitaciones acumuladas')
ax.set_xlabel('Anio')
ax.set_ylabel('Precipitacion acumulada (mm.)')
ax.set_title('Precipitaciones acumuladas segun el anio')
ax.legend()
```

```python
fig.savefig("precipitaciones_anio.pdf")
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.hist(df_lluvias.values.flatten(), bins=10)
ax.set_xlabel(' Histograma Precipitacion')
ax.set_ylabel('Frecuencia Absoluta')
ax.set_title('Histograma Precipitaciones')
```
